```python
x = df[['V','AT','AP']]
y = df['PE']
```

```python
from sklearn.model_selection import train_test_split
x_t,x_test,y_t,y_test = train_test_split(x,y,test_size=0.4)
x_cv,x_train,y_cv,y_train = train_test_split(x_t,y_t,test_size=0.5)
```

```python
x_cv
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import Ridge
cd = Ridge(alpha=20,normalize=True)
cd.fit(x_train,y_train)
```

```python
from sklearn import metrics
metrics.mean_squared_error(cd.predict(x_cv),y_cv)
```

```python
metrics.mean_squared_error(cd.predict(x_train),y_train)
```

```python
m=[]
m1=[]
i=0
while i<200:
    cd=Ridge(alpha=i,normalize=True)
    cd.fit(x_train,y_train)
    m.append(metrics.mean_squared_error(cd.predict(x_train),y_train))
    m1.append(metrics.mean_squared_error(cd.predict(x_cv),y_cv))
    i = i + 1
```

```python
len(m)
```

```python
fig = plt.figure(figsize=(10,6),dpi=100)
plt.plot(np.linspace(1,200,200),m,color='r',label="Train")
plt.plot(np.linspace(1,200,200),m1,color='g',label="CV")
plt.show()
```

```python
import xgboost as xgb
m=[]
m1=[]
i=1
while i<=30:
    cd=xgb.XGBRegressor(max_depth=i)
    cd.fit(x_train,y_train)
    m.append(metrics.mean_squared_error(cd.predict(x_train),y_train))
    m1.append(metrics.mean_squared_error(cd.predict(x_cv),y_cv))
    i = i + 1
```

```python
fig = plt.figure(figsize=(10,6),dpi=100)
plt.plot(np.linspace(1,30,30),m,color='r',label="Train")
plt.plot(np.linspace(1,30,30),m1,color='g',label="CV")
plt.show()
```

```python
m=[]
m1=[]
i=1
while i<=100:
    cd = Ridge(alpha=0.3,normalize=True)
    cd.fit(x_train[0:i],y_train[0:i])
    m.append(metrics.mean_squared_error(cd.predict(x_train[0:i]),y_train[0:i]))
    m1.append(metrics.mean_squared_error(cd.predict(x_cv),y_cv))
    i = i + 1
```

```python
len(m)
```

```python
fig = plt.figure(figsize=(10,6),dpi=100)
plt.plot(np.linspace(1,100,100),m,color='r',label="Train")
plt.plot(np.linspace(1,100,100),m1,color='g',label="CV")
plt.show()
```
